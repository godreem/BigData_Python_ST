# 함수를 이용한 데이터 추출
#inplace #sort_values 

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] < 2:
        return 2
    else:
        return 5    
```

```python
customerData = pd.read_csv("./dataset/customerdata.csv", encoding='ms949')
```

```python
customerData.head()
```

```python
customerData["PRODUCTAGE_ADD"] = customerData.apply(refining, axis=1)
customerData.head()
```

```python
customerData.sort_values(["EMI","AVERAGEPRICE"], ascending=[True,False],inplace=True)
customerData.head()
```
