### 차트 색 및 내용 채우기

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

# 커맨드뷰에서 차트 시연
%matplotlib inline
size = 50
#사이즈만큰 랜덤한 데이터 생성 (정규분포: 평균0, 표준편차1)
y=np.random.standard_normal(size)
x=range(size)
# 차트 꾸미기
plt.plot(x,y, 'b', lw=1.5) #차트 선색 및 크기설정
plt.plot(x,y, 'o', ) #차트 내 'o' 표기
plt.grid(True) #그리드 표기
plt.xlabel('index') # x축 라벨
plt.ylabel('value') # y축 라벨
plt.title('Simple plot') # 차트 제목
```

### 2차원 차트 그리기

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

#첫번째 컬럼자료를 첫번째(왼쪽) 축에 대응하여 그림
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw=1.5, label = '1st')
plt.plot(y[:, 0], 'bo')
plt.grid(True)
plt.legend(loc=4)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
#두번째 컬럼자료를 두번쨰(오른쪽) 축에 대응하여 그림
ax2 = ax1.twinx()
plt.plot(y[:, 1], 'g', lw=1.5, label = '2nd')
plt.plot(y[:, 1], 'go')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

### 2차원 Subplot

```python
#필요 라이브러리 정의
import numpy as np
import matplotlib.pyplot as plt

#데이터 정의 첫번째 컬럼은 정규분포, 두번째 컬럼은 *100
y=np.random.standard_normal((20,2))
y[:,1] = y[:,1]*100

plt.figure(figsize=(7,5))
plt.subplot(211)
plt.plot(y[:, 0], 'b',lw=1.5, label = '1st')
plt.plot(y[:, 0], 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

### 분포도 차트 구현

```python
#pip install seasborn
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
sns.distplot(tips["total_bill"])
```

### jointplot구현

```python
sns.jointplot(x="total_bill", y = "tip", data = tips)
```

### pair plot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = "sex")
```

### barplot 차트

```python
sns.barplot(x='sex', y='total_bill', data = tips)
```

### count plot

```python
sns.countplot(x= 'sex', data = tips)
```

### boxplot 구현`

```python
sns.boxplot(x='day', y='total_bill',data=tips)
```
