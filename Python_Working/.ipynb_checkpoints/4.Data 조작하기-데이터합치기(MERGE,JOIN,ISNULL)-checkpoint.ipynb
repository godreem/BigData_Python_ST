```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
regionMasterData = pd.read_csv("./dataset/kopo_region_mst.csv", encoding='ms949')
```

```python
regionMasterData.head()
```

```python
channelSeasonalityNewData = pd.read_csv("./dataset/kopo_channel_seasonality_new.csv", encoding='ms949')
channelSeasonalityNewData.head()
```

## 데이터 조인

```python
# how 파라미터 기본값은 inner
mergeResult = pd.merge(channelSeasonalityNewData, regionMasterData,\
    left_on="REGIONID", right_on="REGIONID", how="inner")\
    [["REGIONID", "REGIONNAME","PRODUCT","YEARWEEK","QTY"]]


# 결과 건수
#len(mergeResult)

# 결과
mergeResult
```

```python
# how 파라미터 기본값은 inner
mergeResult = pd.merge(channelSeasonalityNewData, regionMasterData,\
    left_on="REGIONID", right_on="REGIONID", how="left")\
    [["REGIONID", "REGIONNAME","PRODUCT","YEARWEEK","QTY"]]


# 결과 건수
len(mergeResult)

# 결과
#mergeResult
```

```python
mergeResult[mergeResult["REGIONNAME"].isnull()]
```
